```python
# To read the file dev.out and represent in suitable format
file = open(r"C:\Users\Matthew\Desktop\SUTD\Term 6\Machine Learning\ML Project\EN\train", "r").read()
file = file.split("\n")
readable = [("Tweet", "START")]
for word in file[:-1]:
    if word != '':
        word = word.split(" ")
        readable.append((word[0], word[1]))
    else:
        readable.append(("/Tweet", "STOP"))
        readable.append(("Tweet", "START"))
readable.pop(-1)

# for read in readable:
#         print(read[0], read[1])
```

```python
# Function estimates transition based on parameters given: tag of word, tag of next word
def trans_param_est(y, y_next):
    counter_y = 0 # To be denominator of MLE equation
    counter_y_next = 0 # To be numerator of MLE equation
    for i in range(len(readable)):
        if readable[i][1] == y:
            counter_y += 1
            if readable[i+1][1] == y_next:
                counter_y_next += 1
    return float(counter_y_next) / float(counter_y)

tags = ["START", "O", "B-negative", "B-positive", "B-neutral", "I-negative", "I-positive", "I-neutral", "STOP"]
```

```python
trans_param = {}
for i in range(len(tags)-1):
    trans_param[tags[i]] = {}
    for j in range(len(tags)):
        if i != j:
            trans_param[tags[i]][tags[j]] = trans_param_est(tags[i], tags[j])
            print("Transition parameters of '%s' to '%s': %f" % (tags[i], tags[j], trans_param_est(tags[i], tags[j])))

# print(trans_param)
```

```python
def viterbi_algorithm(word, state_before, emission_params, trans_params):
    value = 0
    max_value = 0
    max_tag = ""
    if word in emission_params.keys():
        for tag in emission_params[word].keys():
            value = emission_params[word][tag] * trans_params[state_before][tag]
            if value > max_value:
                max_value = value
                max_tag = tag
    return max_tag            
```

```python
# For parsing the text file #
list_of_tweets = []
for sentence in test_file:
    tweet = ["START"]
    words = sentence.split("\n") 
    for word in words:
        tweet.append(word)
    tweet.append("STOP")
    list_of_tweets.append(tweet)

print(list_of_tweets)
```

```python
# For writing the file
output_file = open(r"C:\Users\Matthew\Desktop\SUTD\Term 6\Machine Learning\ML Project\EN\dev.p3.out", "w")
state_before = ""
predicted_state = ""
for tweet in list_of_tweets:
    for word in tweet:
        if word == "START":
            state_before = "START"
        if word != "STOP":
            if word in emission_params.keys():
                predicted_state = viterbi_algorithm(word, state_before, emission_params, trans_param)
                output_file.write(word + " " + predicted_state + "\n")
            else:
                output_file.write("#UNK" + "\n")
        if word == "STOP":
            output_file.write("\n")
output_file.close()
```
