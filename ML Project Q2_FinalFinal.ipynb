```python
import codecs
EN_train = "/Users/aasthachouhan/Desktop/Term 6/Machine Learning /Project/EN/train"
EN_test = "/Users/aasthachouhan/Desktop/Term 6/Machine Learning /Project/EN/dev.in"
EN_output = "/Users/aasthachouhan/Desktop/Term 6/Machine Learning /Project/EN/dev.p2.out"
EN_gold = "/Users/aasthachouhan/Desktop/Term 6/Machine Learning /Project/EN/dev.out"
EN_modified = "/Users/aasthachouhan/Desktop/Term 6/Machine Learning /Project/EN/modified_train"

CN_train = "/Users/aasthachouhan/Desktop/Term 6/Machine Learning /Project/CN/train"
CN_test = "/Users/aasthachouhan/Desktop/Term 6/Machine Learning /Project/CN/dev.in"
CN_output = "/Users/aasthachouhan/Desktop/Term 6/Machine Learning /Project/CN/dev.p2.out"
CN_gold = "/Users/aasthachouhan/Desktop/Term 6/Machine Learning /Project/CN/dev.out"
CN_modified = "/Users/aasthachouhan/Desktop/Term 6/Machine Learning /Project/CN/modified_train"

SG_train = "/Users/aasthachouhan/Desktop/Term 6/Machine Learning /Project/SG/train"
SG_test = "/Users/aasthachouhan/Desktop/Term 6/Machine Learning /Project/SG/dev.in"
SG_output = "/Users/aasthachouhan/Desktop/Term 6/Machine Learning /Project/SG/dev.p2.out"
SG_gold = "/Users/aasthachouhan/Desktop/Term 6/Machine Learning /Project/SG/dev.out"
SG_modified = "/Users/aasthachouhan/Desktop/Term 6/Machine Learning /Project/SG/modified_train"

FR_train = "/Users/aasthachouhan/Desktop/Term 6/Machine Learning /Project/FR/train"
FR_test = "/Users/aasthachouhan/Desktop/Term 6/Machine Learning /Project/FR/dev.in"
FR_output = "/Users/aasthachouhan/Desktop/Term 6/Machine Learning /Project/FR/dev.p2.out"
FR_gold = "/Users/aasthachouhan/Desktop/Term 6/Machine Learning /Project/FR/dev.out"
FR_modified = "/Users/aasthachouhan/Desktop/Term 6/Machine Learning /Project/FR/modified_train"


def modified_training_set(train_file,modified_train_set):
    #fout=open ('modified_train_set','w')
    with open (train_file, encoding='utf-8') as file:
        tag_count={}
        modified_words=[]
        for line in file:
            pair=line.split()
            if len(line.split())!=0:
                tag=pair[0]
                observ=pair[1]
                if tag in tag_count.keys():
                    tag_count[tag]+=1
                else:
                    tag_count[tag]=1
        for tag in tag_count:
            if tag_count[tag]<3:
                modified_words.append(tag)
    with open (train_file, encoding='utf-8') as file2, codecs.open(modified_train_set, 'w', 'utf-8-sig') as fout:
        for line in file2:
            pair2=line.split()
            if len(line.split())!=0:
                word=pair2[0]
                sentiment=pair2[1]
                if word in modified_words:
                    fout.write("#UNK"+" "+sentiment+"\n")
                else:
                    fout.write(word+" "+sentiment+"\n")

def emission_params(train_file):
    with open(train_file, encoding = 'utf-8') as file:
        emission_count= {}
        label_count={}
        for line in file:
            pair = line.split()
            if len(line.split())!=0:
                word = pair[0]
                sentiment = pair[1]
                if word!="#UNK":
                    if word in emission_count.keys():
                        if sentiment in emission_count[word].keys():
                            emission_count[word][sentiment] +=1
                        else:
                            sentiments = emission_count[word]
                            sentiments[sentiment] = 1
                    else:
                        sentiment_count = {}
                        sentiment_count[sentiment] = 1
                        emission_count[word]=sentiment_count
    
                #add 1 to count of label Yi
                    if sentiment in label_count.keys():
                        label_count[sentiment]+=1
                    else:
                        label_count[sentiment]=1
        for keya in emission_count.keys():
            for keyb in emission_count[keya].keys():
                emission_count[keya][keyb]/=(label_count[keyb]+1)
    
    print(emission_count)   
    return (emission_count,label_count)

    
def sentiment_analysis(test_file,output_file,emission_params, observ_count):
    with open(test_file, encoding ='utf-8') as ifile, codecs.open(output_file, 'w', 'utf-8-sig') as ofile:
        for line in ifile:
            if len(line.split())!=0:
                tag = line.split()[0]
                if tag in emission_params.keys():
                    value = emission_params[tag]
                    a = max(value,key=value.get)
                    ofile.write(tag+" "+a+'\n')
                else:
                    ofile.write("#UNK"+" "+a+'\n')
            else:
                ofile.write('\n')
                
                
modified_training_set(EN_train, EN_modified)                
emission_params_EN, label_count_EN = emission_params(EN_modified)
sentiment_analysis(EN_test,EN_output,emission_params_EN, label_count_EN)

modified_training_set(FR_train, FR_modified)                
emission_params_FR, label_count_FR = emission_params(FR_modified)
sentiment_analysis(FR_test,FR_output,emission_params_FR, label_count_FR)

modified_training_set(CN_train, CN_modified)                
emission_params_CN,label_count_CN = emission_params(CN_modified)
sentiment_analysis(CN_test,CN_output,emission_params_CN, label_count_CN)

modified_training_set(SG_train, SG_modified)                
emission_params_SG,label_count_SG = emission_params(SG_modified)
sentiment_analysis(SG_test,SG_output,emission_params_SG, label_count_SG)

#EN
#Entity in gold data: 226
#Entity in prediction: 773
#Correct Entity : 136
#Entity  precision: 0.1759
#Entity  recall: 0.6018
#Entity  F: 0.2723
#Correct Sentiment : 40
#Sentiment  precision: 0.0517
#Sentiment  recall: 0.1770
#Sentiment  F: 0.0801

#CN
#Entity in gold data: 362
#Entity in prediction: 2015
#Correct Entity : 80
#Entity  precision: 0.0397
#Entity  recall: 0.2210
#Entity  F: 0.0673
#Correct Sentiment : 45
#Sentiment  precision: 0.0223
#Sentiment  recall: 0.1243
#Sentiment  F: 0.0379

#FR
#Entity in gold data: 223
#Entity in prediction: 846
#Correct Entity : 173
#Entity  precision: 0.2045
#Entity  recall: 0.7758
#Entity  F: 0.3237
#Correct Sentiment : 65
#Sentiment  precision: 0.0768
#Sentiment  recall: 0.2915
#Sentiment  F: 0.1216

#SG
#Entity in gold data: 1382
#Entity in prediction: 3171

#Correct Entity : 380
#Entity  precision: 0.1198
#Entity  recall: 0.2750
#Entity  F: 0.1669
#Correct Sentiment : 193
#Sentiment  precision: 0.0609
#Sentiment  recall: 0.1397
#Sentiment  F: 0.0848
```
